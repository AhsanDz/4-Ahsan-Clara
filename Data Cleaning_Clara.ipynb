**Skenario:**
Kamu adalah Lead Data Analyst di sebuah toko gadget besar. Bosmu memberikan dataset spesifikasi smartphone dan memintamu menjawab **10 Pertanyaan Bisnis** untuk strategi penjualan bulan depan.

**Aturan Main:**

```python
# ⚠️ JALANKAN CELL INI TERLEBIH DAHULU!
# Script ini akan membuat dataset dummy untuk latihan kita hari ini.

import pandas as pd
import numpy as np
import random

np.random.seed(42)

# Data Dummy
n_data = 150
brands = ['Samsung', 'Apple', 'Xiaomi', 'Oppo', 'Vivo', 'Realme', 'Infinix', 'Asus', 'Google']
rams = [4, 6, 8, 12, 16, 24] # Ada RAM 24GB buat gaming
storages = [64, 128, 256, 512, 1024]

data = {
    'Brand': [random.choice(brands) for _ in range(n_data)],
    'Model': [f"Phone {i}" for i in range(n_data)],
    'RAM_GB': [random.choice(rams) for _ in range(n_data)],
    'Storage_GB': [random.choice(storages) for _ in range(n_data)],
    'Rating': [round(random.uniform(3.0, 5.0), 1) for _ in range(n_data)]
}

df = pd.DataFrame(data)

# Bikin harga agak logis -> (RAM gede + Storage gede = Mahal) + sedikit random noise
df['Price_Juta'] = (df['RAM_GB'] * 0.5) + (df['Storage_GB'] * 0.015) + np.random.randint(1, 4, n_data)
df['Price_Juta'] = df['Price_Juta'].round(1)

# Tambahin Data -> "Aneh" (Anomaly)
df.loc[148] = ['Nokia', 'Nokia 3310 Reborn', 2, 16, 50.0, 5.0] # HP kentang harga sultan
df.loc[149] = ['Unknown', 'Prototype X', 128, 2048, 0.5, 1.0] # Spek dewa harga permen

print("Dataset 'smartphones' berhasil dibuat!")
print(f"Total Data: {len(df)} baris.")
df.head(10)
```

### **Quest 2: Raja Pasar**
Brand apa yang paling banyak memiliki model HP dalam daftar ini?

*Hint: Gunakan `.value_counts()`.*

```python
data = df['Brand'].value_counts()
print(data)
max = data.max()
print(f'Brand dengan jumlah model HP terbanyak dengan jumlah {max} buah adalah', data.idxmax())
```

Untuk mengetahui brand dengan model terbanyak, digunakan function .value_counts() pada kolom brand. Didapatkan bahwa brand Xiaomi memiliki 21 model. Maka dapat disimpulkan bahwa brand Xiaomi adalah brand dengan model HP terbanyak dari data tersebut.

### **Quest 3: Storage Wars**
Zaman sekarang aplikasi makin besar.
Coba cari HP yang punya **Storage paling besar** (Max Storage). Ada berapa HP yang punya storage segitu?

*Hint: Cari nilai max dulu, lalu filter datanya.*

```python
data_real = df[df['Storage_GB'].isin(storages)]
max_storage = data_real['Storage_GB'].max()
print('Storage Paling Besar:', max_storage, 'GB')

a = data_real['Storage_GB'].value_counts()
print('Berikut adalah rincian jumlah brand HP berdasarkan penyimpanannya: ')
print(a)

jumlah_1024 = a[max_storage]
print()
print(f'Terdapat {jumlah_1024} HP dengan storage sebesar {max_storage} GB')
list_brand = data_real[data_real['Storage_GB'] == max_storage]
print(f'Berikut adalah rincian brand HP dengan penyimpanan terbesar, yakni sebanyak {jumlah_1024} buah')
print(list_brand)
```

Berdasarkan cell di atas, diketahui bahwa ada data anomali. Maka, kita melakukan filter data dengan membuat variabel `data_real` dan menggunakan function .isin() untuk menampung data-data yang sesuai dengan nilai storages.

Selanjutnya, cell ini dikerjakan dengan menggunakan variabel data_real. Didapatkan juga detail jumlah HP dengan penyimpanan tertentu.

Setelah itu, kita melakukan filter untuk mencari data-data HP yang memiliki penyimpanan 1024 GB. Kemudian data tersebut dicetak

### **Quest 4: Budget vs Premium**
Bos ingin tahu rentang harga pasar.
1. Berapa harga HP **termurah**?
2. Berapa harga HP **termahal**?

*Hint: Gunakan `.min()` dan `.max()` pada kolom harga.*

```python
# Data anomali belum disaring

min = df['Price_Juta'].min()
max = df['Price_Juta'].max()
print('Harga HP termurah:', min, 'juta')
print('Harga HP termahal:', max, 'juta')

# Kode dengan data anomali yang sudah disaring

data_real = df[
    ~((df['RAM_GB'] < 4) & (df['Price_Juta'] > 10)) & # Hapus jika RAM < 4 DAN Harga > 10
    ~((df['RAM_GB'] > 32) & (df['Price_Juta'] < 5)) & # Hapus jika RAM > 32 DAN Harga < 5
    ~(df['Rating'] > 5)
]

urutan_harga = data_real['Price_Juta'].value_counts()
min_baru = data_real['Price_Juta'].min()
max_baru = data_real['Price_Juta'].max()
tmp_min = urutan_harga[min_baru]
tmp_max = urutan_harga[max_baru]
list_min = data_real[data_real['Price_Juta'] == min_baru]
list_max = data_real[data_real['Price_Juta'] == max_baru]
print()
print(f'Harga HP termurah dengan data baru: {min_baru} juta. Terdapat {tmp_min} HP dengan harga {min_baru} juta, yakni:')
print(list_min)
print()
print(f'Harga HP termahal dengan data baru: {max_baru} juta. Terdapat {tmp_max} HP dengan harga {max_baru} juta, yakni:')
print(list_max)
```

Kode paling atas berfungsi untuk menemukan data harga HP termurah dan harga HP termahal dengan data anomali yang belum disaring

Sementara itu, kode di bawah berfungsi untuk menemukan data harga HP termurah dan termahal dengan data anomali yang sudah disaring. data_real adalah variabel yang digunakan untuk menampung data yang bersih. Maka dari itu, kode dibawahnya selalu menggunakan variabel ini.

### **Quest 5: The Gaming Beast**
Seorang Gamer datang mencari HP. Syaratnya mutlak:
1. **RAM wajib di atas 12 GB.**
2. **Storage minimal 512 GB.**

Tampilkan daftar HP yang memenuhi syarat tersebut!

*Hint: Gunakan Filtering dengan logika AND (`&`).*

```python
data_bersyarat = data_real[
    (data_real['RAM_GB'] > 12) & (data_real['Storage_GB'] >= 512)
]
total = len(data_bersyarat)

print(f'Terdapat {total} HP yang memenuhi syarat untuk gamer tersebut. Diantaranya adalah: ')
print(data_bersyarat)
```

Untuk mendapatkan data HP sesuai syarat tersebut, digunakan logika AND. Disini, digunakan data_real karena variabel ini menyimpan data yang sudah disaring di cell sebelumnya.

Kemudian, variabel total digunakan untuk menampung jumlah HP yang memenuhi syarat dan daftar HP tersebut dicetak.

### **Quest 6: Sobat Mahasiswa**
Mahasiswa biasanya cari HP yang "Worth it".
Cari HP yang harganya **di bawah 7 Juta**, TAPI **Rating-nya di atas 4.5**.

Ada nggak barangnya?

```python
hp_mahasiswa = data_real[
    (data_real['Price_Juta'] < 7) & (data_real['Rating'] > 4.5)
]
jumlah_hp = len(hp_mahasiswa)

print(f'Terdapat {jumlah_hp} HP yang sesuai dengan kriteria mahasiswa. Diantaranya adalah: ')
print(hp_mahasiswa)
```

Untuk menjalankan kode ini, digunakan variabel data_real yang sudah bersih dari data anomali.

Variabel hp_mahasiswa digunakan untuk menyimpan data dengan kriteria harga di bawah 7 juta dan rating di atas 4.5. Maka dari itu, digunakan operasi AND (&) agar kedua kondisi terpenuhi

Menggunakan len(), variabel jumlah_hp digunakan untuk menampung jumlah HP yang sesuai dengan kriteria tersebut. Kemudian hp_mahasiswa dicetak.

### **Quest 7: Fanboy Apple**
Ada pelanggan yang cuma mau beli **Apple**.
Tampilkan semua HP merk Apple, lalu urutkan dari yang **paling mahal ke paling murah**.

*Hint: Filter `Brand == 'Apple'`, lalu `sort_values`.*

```python
data_apple = data_real[data_real['Brand'] == 'Apple']
harga_apple = data_apple.sort_values(by='Price_Juta', ascending=False)
jumlah_apple = len(data_apple)
print(f'Terdapat {jumlah_apple} HP merk Apple. Berikut adalah harga HP Apple yang telah diurutkan: ')
print(harga_apple)
```

data_apple digunakan untuk menampung data HP dengan brand Apple.

harga_apple digunakan untuk menampung urutan HP berdasarkan harga mulai harga paling mahal ke murah. `data_apple.sort_values(by='Price_Juta', ascending=False)` adalah syntax untuk menyaring kolom Price_Juta dari terbesar ke terkecil. Jika tidak ditambahkan ascending=false, maka variabel ini menampung data dengan urutan data termurah ke termahal.

### **Quest 8: Rata-Rata Harga per Brand (Level Medium)**
Brand mana yang rata-rata harga produknya paling mahal?
Coba hitung rata-rata harga (`Price_Juta`) untuk setiap `Brand`.

*Hint: Ini butuh Grouping. Gunakan `.groupby('Brand')['Price_Juta'].mean()`.*

```python
rata_rata_harga = data_real.groupby('Brand')['Price_Juta'].mean()
rata_rata_terbesar = rata_rata_harga.max()
brand_termahal = rata_rata_harga.idxmax()
print(rata_rata_harga)
print(f'Brand dengan rata-rata harga produk paling mahal adalah {brand_termahal} dengan rata-rata mencapai {rata_rata_terbesar} juta')
```

Dengan menggunakan function .groupby('Brand')['Price_Juta'].mean(), kita membuat series dengan Brand sebagai indexnya. Dengan menggunakan .mean(), kita menghitung Price_Juta pada index yang sama.

Dengan .max() dan .idxmax(), kita mengambil nilai terbesar dan index terbesar dari series tersebut, kemudian dicetak.

### **Quest 9: Value for Money (Feature Engineering)**
Kita mau cari HP yang "Spek Dewa Harga Kaki Lima".
1. Buat kolom baru bernama `Skor_Performa` = `RAM_GB` + `Storage_GB`.
2. Buat kolom baru bernama `Value_Skor` = `Skor_Performa` / `Price_Juta`.
3. Tampilkan 5 HP dengan `Value_Skor` tertinggi.

*Hint: Kamu bisa melakukan operasi matematika antar kolom secara langsung.*

```python
data_real['Skor_Performa'] = data_real['RAM_GB'] + data_real['Storage_GB']
data_real['Value_Skor'] = data_real['Skor_Performa'] / data_real['Price_Juta']
data_spekdewa = data_real.sort_values(by='Value_Skor', ascending=False)
print(data_spekdewa.head())
```

Kita membuat kolom baru bernama Skor_Performa dan Value_Skor. Hal ini perlu dilakukan karena kita akan melakukan sortir pada kolom Value_Skor. Dengan menambahkan ascending=False, data telah diurutkan dari yang terbesar ke yang terkecil.

5 data teratas dicetak menggunakan .head()

Muncul peringatan setelah kode dijalankan. Hal ini dikarenakan kita bekerja pada slice dari data yang asli.

### **Quest 10: The Detective (Final Boss)**
Ada laporan data error di sistem.
1. Cari HP yang harganya **di atas 40 Juta** (Kemahalan/Mustahil).
2. Cari HP yang harganya **di bawah atau sama dengan 1 Juta** tapi RAM-nya **di atas 100 GB** (Ngawur).

Temukan data penyusup ini!

*Hint: Gunakan filtering untuk mencari outlier.*

```python
data_penyusup = df[(df['Price_Juta'] > 40) | ((df['Price_Juta'] <= 1) & (df['RAM_GB'] > 100))]
kolom = ['Brand', 'Model', 'RAM_GB', 'Storage_GB', 'Rating', 'Price_Juta']
print(data_penyusup[kolom])
```

Untuk menemukan data penyusup ini, kita tidak bisa menggunakan data_real. Hal ini dikarenakan data_real telah disortir sehingga tidak ada data anomali/penyusupnya.

Digunakan AND dan OR untuk menemukan data penyusup sesuai kriteria di atas. Dikarenakan kita telah melakukan .copy pada data di atas, maka kolom yang kita buat pada cell sebelumnya ikut terbuat. Maka dari itu, kita hanya membutuhkan kolom-kolom yang relevan dengan data yang kita butuhkan.
